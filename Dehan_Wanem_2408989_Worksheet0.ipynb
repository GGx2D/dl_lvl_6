# **Task1**

Create a Python program that converts between different units of measurement.
• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.

• Requirements:
1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and
return value.


• Conversion Options:
1. Length:

    – Convert meters (m) to feet (ft).

    – Convert feet (ft) to meters (m).

2. Weight:


    – Convert kilograms (kg) to pounds (lbs).

    – Convert pounds (lbs) to kilograms (kg).

3. Volume:


    – Convert liters (L) to gallons (gal).

    – Convert gallons (gal) to liters (L).

```python
def convert_units(category, value, choice):
    """
    Convert units based on category and choice.

    category: length / weight / volume
    value: number to convert
    choice: "1" or "2"
    returns: converted number
    """
    if category == "length":
        if choice == "1":      # meters -> feet
            return value * 3.28084
        elif choice == "2":    # feet -> meters
            return value / 3.28084

    elif category == "weight":
        if choice == "1":      # kg -> pounds
            return value * 2.20462
        elif choice == "2":    # pounds -> kg
            return value / 2.20462

    elif category == "volume":
        if choice == "1":      # liters -> gallons
            return value * 0.264172
        elif choice == "2":    # gallons -> liters
            return value / 0.264172

    raise ValueError("Invalid choice.")


def unit_converter():
    """
    Ask user for conversion type and value, then print the result.
    Handles invalid category, invalid number input, and wrong choice.
    """
    print("Choose category: length / weight / volume")
    category = input("Category: ").lower().strip()

    # check category to ensure its there
    if category not in ["length", "weight", "volume"]:
        print("Invalid category.")
        return

    try:
        value = float(input("Enter value: "))
    except ValueError:
        print("Invalid input! Please enter a number.")
        return

    if category == "length":
        print("1 = meters to feet")
        print("2 = feet to meters")
    elif category == "weight":
        print("1 = kg to pounds")
        print("2 = pounds to kg")
    elif category == "volume":
        print("1 = liters to gallons")
        print("2 = gallons to liters")

    choice = input("Choose 1 or 2: ").strip()

    try:
        result = convert_units(category, value, choice)
    except ValueError as e:
        print("Error:", e)
        return

    # Display result
    if category == "length" and choice == "1":
        print("Feet =", result)
    elif category == "length" and choice == "2":
        print("Meters =", result)
    elif category == "weight" and choice == "1":
        print("Pounds =", result)
    elif category == "weight" and choice == "2":
        print("Kg =", result)
    elif category == "volume" and choice == "1":
        print("Gallons =", result)
    elif category == "volume" and choice == "2":
        print("Liters =", result)


unit_converter()
```

# ***Task-2***

Create a Python program that performs various mathematical operations on a list of numbers.

• The Program should:
1. Prompt the user to choose an operation (e.g., find the sum, average, maximum, or minimum
of the numbers).
2. Ask the user to input a list of numbers (separated by spaces).

3. Perform the selected operation and display the result.
4. Handle potential errors, such as invalid input or empty lists.


• Requirements:

1. Functions: Define at least one function for each operation (sum, average, maximum, mini-
mum).

2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values or
empty lists).
3. User Input: Prompt the user to select the operation and input the list of numbers.
4. Docstrings: Include a docstring in each function to describe its purpose, parameters, and
return value.


```python
def sum_numbers(nums):
    """Return the sum of a list of numbers."""
    total = 0
    for x in nums:
        total += x
    return total

def average_numbers(nums):
    """Return the average of a list of numbers."""
    if len(nums) == 0:
        raise ValueError("Empty list")
    return sum_numbers(nums) / len(nums)

def max_number(nums):
    """Return the maximum value in a list of numbers."""
    if len(nums) == 0:
        raise ValueError("Empty list")
    m = nums[0]
    for x in nums:
        if x > m:
            m = x
    return m

def min_number(nums):
    """Return the minimum value in a list of numbers."""
    if len(nums) == 0:
        raise ValueError("Empty list")
    m = nums[0]
    for x in nums:
        if x < m:
            m = x
    return m

def list_math_program():
    """Ask user for an operation and numbers, then print the result."""
    print("Choose operation: sum / avg / max / min")
    op = input("Operation: ").lower().strip()

    if op not in ["sum", "avg", "max", "min"]:
        print("Invalid operation.")
        return

    try:
        parts = input("Enter numbers separated by spaces: ").split()
        if len(parts) == 0:
            print("Empty list!")
            return
        nums = [float(x) for x in parts]
    except ValueError:
        print("Invalid input! Please enter only numbers.")
        return

    if op == "sum":
        print("Sum =", sum_numbers(nums))
    elif op == "avg":
        print("Average =", average_numbers(nums))
    elif op == "max":
        print("Max =", max_number(nums))
    elif op == "min":
        print("Min =", min_number(nums))

list_math_program()
```

# **4.3 Exercise on Nested List:**

**1. Flatten a Nested List:**

Write a Python function that takes a nested list and flattens it into a single list, where all the elements
are in a single dimension.

*• Requirements:*

    – Define a function flatten(lst) that takes a nested list lst and returns a flattened version
    of the list.

    – Example: For the input [[1, 2], [3, 4], [5]], the output should be [1, 2, 3, 4, 5].

```python
def flatten(lst):
    """Flatten a list of lists into one list (one level)."""
    out = []
    for sub in lst:
        for item in sub:
            out.append(item)
    return out

print(flatten([[1, 2], [3, 4], [5]]))
```

**2. Accessing Nested List Elements:**

Write a Python function that extracts a specific element from a nested list given its indices.

*• Requirements:*

    – Define a function access nested element(lst, indices) that takes a nested list lst and
    a list of indices indices, and returns the element at that position.

    – Example: For the input lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]] with indices = [1,
    2], the output should be 6.

```python
def access_nested_element(lst, indices):
    """Return element from nested list using a list of indices."""
    current = lst
    for idx in indices:
        current = current[idx]
    return current

lst = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(access_nested_element(lst, [1, 2]))
```

**3. Sum of All Elements in a Nested List:**

Write a Python function that calculates the sum of all the numbers in a nested list (regardless of depth).

• Requirements: *italicized text*

    – Define a function sum nested(lst) that takes a nested list lst and returns the sum of all
    the elements.

    – Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 21.

```python
def sum_nested(lst):
    """Return sum of all numbers in a nested list."""
    total = 0
    for item in lst:
        if type(item) == list:
            total += sum_nested(item)
        else:
            total += item
    return total

print(sum_nested([[1, 2], [3, [4, 5]], 6]))
```

**4. Remove Specific Element from a Nested List:**

Write a Python function that removes all occurrences of a specific element from a nested list.

*• Requirements:*

    – Define a function remove element(lst, elem) that removes elem from lst and returns the
    modified list.

    – Example: For the input lst = [[1, 2], [3, 2], [4, 5]] and elem = 2, the output should
    be [[1], [3], [4, 5]].

```python
def remove_element(lst, elem):
    """Remove all occurrences of elements from nested list (any depth)."""
    out = []
    for item in lst:
        if type(item) == list:
            out.append(remove_element(item, elem))
        else:
            if item != elem:
                out.append(item)
    return out

print(remove_element([[1, 2], [3, 2], [4, 5]], 2))
```

**5. Find the Maximum Element in a Nested List:**

Write a Python function that finds the maximum element in a nested list (regardless of depth).

*• Requirements:*

    – Define a function find max(lst) that takes a nested list lst and returns the maximum
    element.

    – Example: For the input [[1, 2], [3, [4, 5]], 6], the output should be 6.

```python
def find_max(lst):
    """Return the maximum element in a nested list (any depth)."""
    current_max = None
    for item in lst:
        if type(item) == list:
            val = find_max(item)
        else:
            val = item

        if current_max is None or val > current_max:
            current_max = val
    return current_max

print(find_max([[1, 2], [3, [4, 5]], 6]))
```

**6. Count Occurrences of an Element in a Nested List:**

Write a Python function that counts how many times a specific element appears in a nested list.

*• Requirements:*

    – Define a function count occurrences(lst, elem) that counts the occurrences of elem in
    the nested list lst.

    – Example: For the input lst = [[1, 2], [2, 3], [2, 4]] and elem = 2, the output should
    be 3.

```python
def count_occurrences(lst, elem):
    """Count how many times elements appears in nested list (any depth)."""
    count = 0
    for item in lst:
        if type(item) == list:
            count += count_occurrences(item, elem)
        else:
            if item == elem:
                count += 1
    return count

print(count_occurrences([[1, 2], [2, 3], [2, 4]], 2))
```

**7. Flatten a List of Lists of Lists:**

Write a Python function that flattens a list of lists of lists into a single list, regardless of the depth.

*• Requirements:*

    – Define a function deep flatten(lst) that takes a deeply nested list lst and returns a single
    flattened list.

    – Example: For the input [[[1, 2], [3, 4]], [[5, 6], [7, 8]]], the output should be
    [1, 2, 3, 4, 5, 6, 7, 8].

```python
def deep_flatten(lst):
    """Flatten a nested list of any depth into one list."""
    out = []
    for item in lst:
        if type(item) == list:
            out.extend(deep_flatten(item))
        else:
            out.append(item)
    return out

print(deep_flatten([[[1, 2], [3, 4]], [[5, 6], [7, 8]]]))
```

**8. Nested List Average:**

Write a Python function that calculates the average of all elements in a nested list.

*• Requirements:*

    – Define a function average nested(lst) that takes a nested list lst and returns the average
    of all the elements.

    – Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
def average_nested(lst):
    """Return average of all numbers in nested list (any depth)."""
    flat = deep_flatten(lst)
    return sum(flat) / len(flat)

print(average_nested([[1, 2], [3, 4], [5, 6]]))  # 3.5
```

**• Solve the following system of Linear equation using Inverse Methods.**

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}

**2x − 3y + z = −1**

**x − y + 2z = −3**

**3x + y − z = 9**

```python
import numpy as np

A = np.array([[2, -3,  1],
              [1, -1,  2],
              [3,  1, -1]], dtype=float)

B = np.array([-1, -3, 9], dtype=float)

X = np.linalg.solve(A, B)

print("x =", X[0])
print("y =", X[1])
print("z =", X[2])
```

Now: solve the above equation using np.linalg.inv function. {Explore more about ”linalg” function of Numpy}

```python
#coefficients
A = np.array([[2, -3,  1],
              [1, -1,  2],
              [3,  1, -1]], dtype=float)

#constants
B = np.array([-1, -3, 9], dtype=float)

# Inverse method: X = inv(A) * B
X = np.linalg.inv(A) @ B

print("x =", X[0])
print("y =", X[1])
print("z =", X[2])
```

```python
SIZE = 1000

# Python Lists
A = [list(range(SIZE)) for _ in range(SIZE)]
B = [list(range(SIZE)) for _ in range(SIZE)]

def list_matmul(A, B):
    result = [[0] * SIZE for _ in range(SIZE)]
    for i in range(SIZE):
        for k in range(SIZE):
            for j in range(SIZE):
                result[i][j] += A[i][k] * B[k][j]
    return result

start = time.perf_counter()
C = list_matmul(A, B)
end = time.perf_counter()
print("List matrix multiplication time:", end - start, "seconds")
```
